```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

#1. Load the data
data = pd.read_csv('/content/Realisitc Hypothetical Student Sleep Data.csv')


# 2. Prepare input values and target
X = data[['Hours_Studied', 'Slept_Well']].values
y = data['Passed'].values

# 3. Train the logistic regression model
model = LogisticRegression()
model.fit(X, y)

# 4. Generate test data: varying Hours_Studied, fixed Slept_Well = 1
hours = np.linspace(0, 7, 100)
sleep = np.ones_like(hours)  # slept well = 1
X_test = np.column_stack((hours, sleep))

# 5. Predict probabilities using the trained model
prob_pass = model.predict_proba(X_test)[:, 1]  # probability of passing (class 1)

# 6. Then, if you want to plot the sigmoid curve to better visualize the chart, do this.
plt.figure(figsize=(8, 5))
plt.plot(hours, prob_pass, label='Predicted Probability of Passing', color='blue')
plt.scatter(data['Hours_Studied'], data['Passed'], c=data['Slept_Well'], cmap='coolwarm', edgecolors='k', label='Training Data')
plt.xlabel('Hours Studied')
plt.ylabel('Probability of Passing')
plt.title('Predicted Probability of Passing vs. Hours Studied (Assuming Good Sleep)')
plt.grid(True)
plt.legend()
plt.show()
```
